```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./world-happiness-report-2021.csv')
df
```

## Assignment 1
행복 지수를 1점 간격으로 두고 다음과 같은 **Histogram**을 작성 해 주세요.
행복 지수의 키 값은 "Ladder score" 입니다.

```python
plt.hist(df['Ladder score'], range(2,10))
plt.xlabel("score")
plt.ylabel("count")
plt.title("Happiness Score")
plt.ylim(0, 50)  # y 범위 설정 (start, end)
plt.show()
```

## Assignment 2
행복 지수를 y축으로, GDP를 x축으로 하여, scatter 한 값을 한 번 입력 해 보세요.

```python
plt.scatter(df["Logged GDP per capita"], df["Ladder score"])
plt.xlabel("GDP")
plt.ylabel("score")
plt.title("Happiness Score per GDP")
plt.show()
```

## Assignment 3
행복지수 상위 10개, 하위 10개를 선택 후, 각각의 Social support, Logged GDP per capita, Freedom to make life choices 의 평균을 구해서 꺾은 선 그래프로 나타내세요.

```python
best_happy_df = df[:10]
worst_happy_df = df[-10:]

IND = ["Social support" , "Logged GDP per capita" ,"Freedom to make life choices" ]
x = ['Social', 'GDP Attribute', 'Freedom']
y1 = [best_happy_df[IND[i]].mean() for i in range(len(IND))]
y2 = [worst_happy_df[IND[i]].mean() for i in range(len(IND))]
#Scaling for visualizing
y1[0] *= 10
y2[0] *= 10
y1[2] *= 10
y2[2] *= 10


x_axis = np.arange(len(x))
plt.xticks(x_axis, x)

plt.title("Happy Country and Unhappy Country")
plt.ylabel("Score")

plt.bar(x_axis - 0.2, y1, width=0.3, color='green', align='center', label='Happy')
plt.bar(x_axis + 0.2, y2, width=0.3, color='blue', align='center', label='UnHappy')
plt.legend(loc=1)

plt.show()
```

```python
plt.title("Happy Country and Unhappy Country")
plt.ylabel("Score")
plt.plot(x, y1,'g-', label='Happy')
plt.plot(x, y2,'b-', label='UnHappy')
plt.legend(loc=1)
plt.show()
```
