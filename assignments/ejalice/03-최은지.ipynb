```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./world-happiness-report-2021.csv')
df
```

## Assignment 1
행복 지수를 1점 간격으로 두고 다음과 같은 **Histogram**을 작성 해 주세요.
행복 지수의 키 값은 "Ladder score" 입니다.

```python
s=df["Ladder score"]
plt.hist(s, [i for i in range(2, 10)])
plt.xlabel("score")
plt.ylabel("count")
plt.title("Happiness Score")
plt.show()
```

## Assignment 2
행복 지수를 y축으로, GDP를 x축으로 하여, scatter 한 값을 한 번 입력 해 보세요.

```python
y=df["Ladder score"]
x=df["Logged GDP per capita"]
plt.scatter(x, y)
plt.title("Happiness Score per GDP")
plt.xlabel("GDP")
plt.ylabel("score")
plt.show()
```

## Assignment 3
행복지수 상위 10개, 하위 10개를 선택 후, 각각의 Social support, Logged GDP per capita, Freedom to make life choices 의 평균을 구해서 막대 그래프로 나타내세요.

```python
df2=df.sort_values(by="Ladder score", ascending=False)
df2["Social support"]=df2["Social support"]*10
df2["Freedom to make life choices"]=df2["Freedom to make life choices"]*10

Happy=df2.head(10).loc[:,["Social support", "Logged GDP per capita", "Freedom to make life choices"]].mean()
Unhappy=df2.tail(10).loc[:,["Social support", "Logged GDP per capita", "Freedom to make life choices"]].mean()

x=["Social", "GDP Attribute", "Freedom"]

y1=Happy
y2=Unhappy

x_axis=np.arange(len(x))
plt.xticks(x_axis, x)

plt.bar(x_axis-0.2, y1, width=0.3, color="green", align="center", label="Happy")
plt.bar(x_axis+0.2, y2, width=0.3, color="Blue", align="center", label="Unhappy")
plt.legend(loc=1)
plt.ylabel("score")
plt.title("Happy Country and Unhappy Country")
plt.show()
```
