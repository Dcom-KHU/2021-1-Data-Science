생각해 보니 레드 카드도 감안을 해야 할 것 같습니다. 레드 카드 5점, 옐로우 카드 2점으로 점수를 산정 하여, 해당 점수를 바탕으로 정렬해서, 진짜 과격한 리그가 어느 리그인지 알아 보도록 해봐요!

아 맞다, df['A'] = df['B'] * 5 + df['C'] * 2

이런 식으로 새로운 Column을 만들면 되지 않을까요?

```python
grouped=df.groupby('Tournament').sum().loc[:,'yellow_cards':'red_cards'].sort_values(by='yellow_cards',ascending=False)
grouped.assign(score=0)
grouped['score']=grouped['yellow_cards']*2+grouped['red_cards']*5
grouped
```
