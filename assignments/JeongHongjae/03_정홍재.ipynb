## Assignment 1
행복 지수를 1점 간격으로 두고 다음과 같은 **Histogram**을 작성 해 주세요.
행복 지수의 키 값은 "Ladder score" 입니다.

```python
df2 = df["Ladder score"]
plt.hist(df2, [i for i in range(2, 10)], density=True)  # target, int or sequence, density
plt.xlabel("score")
plt.ylabel("count")
plt.show()
```
