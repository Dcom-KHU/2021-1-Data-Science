```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./world-happiness-report-2021.csv')
df.sort_values(by="Ladder score", ascending=False)
```

## Assignment 1
행복 지수를 1점 간격으로 두고 다음과 같은 **Histogram**을 작성 해 주세요.
행복 지수의 키 값은 "Ladder score" 입니다.

```python
plt.hist(df.loc[:, "Ladder score"].astype(int), width=1.5)
plt.xlabel("score")
plt.ylabel("count")
plt.ylim(0, 50)
plt.xlim(0, 9)
plt.title("Happiniess Score")
plt.show()
```

## Assignment 2
행복 지수를 y축으로, GDP를 x축으로 하여, scatter 한 값을 한 번 입력 해 보세요.

```python
x = df.loc[:, "Explained by: Log GDP per capita"]
y = df.loc[:, "Ladder score"]

fig, ax = plt.subplots()
ax.scatter(x, y, c='green')

ax.set_xlabel("GDP")
ax.set_ylabel("score")
ax.set_title("Happiniess Score per GDP", fontsize = 15)

plt.show()
```

## Assignment 3
행복지수 상위 10개, 하위 10개를 선택 후, 각각의 Social support, Logged GDP per capita, Freedom to make life choices 의 평균을 구해서 막대 그래프로 나타내세요.

```python
high = df.loc[0:10, ["Social support", "Logged GDP per capita", "Freedom to make life choices"]]
low = df.loc[139:150, ["Social support", "Logged GDP per capita", "Freedom to make life choices"]]
high_low = pd.concat([high.mean(), low.mean()], axis = 1)
high_low.index = ["Social", "GDP", "Freedom"]
high_low.columns = ["Happy", "Unhappy"]
# print(high_low)
high_low.plot(kind='bar')
plt.xticks(rotation=0, fontsize=13)
plt.show()
```
