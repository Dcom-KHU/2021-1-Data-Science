## Assignment 3
행복지수 상위 10개, 하위 10개를 선택 후, 각각의 Social support, Logged GDP per capita, Freedom to make life choices 의 평균을 구해서 꺾은 선 그래프로 나타내세요.

```python
unhappy = df.sort_values('Ladder score').head(10).mean()
happy = df.sort_values('Ladder score',ascending = False).head(10).mean()

values = ['Social support','Logged GDP per capita','Freedom to make life choices']

df1 = pd.DataFrame(columns = ['happy','unhappy'])
df1['unhappy'] = unhappy[values]
df1['happy'] = happy[values]


df1.plot(kind = 'bar')
```
