```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./world-happiness-report-2021.csv')
df
```

## Assignment 1
행복 지수를 1점 간격으로 두고 다음과 같은 Histogram을 작성 해 주세요. 행복 지수의 키 값은 "Ladder score" 입니다.

```python
s = df.loc[:, 'Ladder score'].values   # df['Ladder score']
plt.hist(s, [i for i in range(2,10)])
plt.xlabel('score')
plt.ylabel('count')
plt.title('Happiness Score')
plt.show()
```

## Assignment 2
행복 지수를 y축으로, GDP를 x축으로 하여, scatter 한 값을 한 번 입력 해 보세요.

```python
x = df.loc[:, 'Logged GDP per capita'].values
y = df.loc[:, 'Ladder score'].values
plt.scatter(x, y)
plt.xlabel('GDP')
plt.ylabel('score')
plt.title('Happiness Score per GDP')
plt.show()
```

## Assignment 3
행복지수 상위 10개, 하위 10개를 선택 후, 각각의 Social support, Logged GDP per capita, Freedom to make life choices 의 평균을 구해서 막대 그래프로 나타내세요.

```python
x = ['Social', 'GDP', 'Freedom']
x_axis = np.arange(len(x))
plt.xticks(x_axis, x)

y1_head10 = df.sort_values(by='Ladder score', ascending=False).head(10)
y2_tail10 = df.sort_values(by='Ladder score', ascending=False).tail(10)
y1 = [y1_head10['Social support'].mean()*10, y1_head10['Logged GDP per capita'].mean(), y1_head10['Freedom to make life choices'].mean()*10]
y2 = [y2_tail10['Social support'].mean()*10, y2_tail10['Logged GDP per capita'].mean(), y2_tail10['Freedom to make life choices'].mean()*10]

plt.bar(x_axis-0.2, y1, width=0.3, color='green', label='Happy')
plt.bar(x_axis+0.2, y2, width=0.3, color='blue', label='Unhappy')

plt.xlabel('Attribute')
plt.ylabel('Score')
plt.title('Happy Country and Unhappy Country')
plt.legend(loc=1)
plt.show()
```
